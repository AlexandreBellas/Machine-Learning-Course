# Python

Este é um mini-tutorial para demonstrar algumas das principais diferenças de Python e outras linguagens como C e Java. Ele é destinado para pessoas que já tem algum contato com outras linguagens de programação. Está é uma adaptação do tutorial que pode ser encontrado [aqui](https://nerdparadise.com/programming/python4minutes/).



## Explicação da sintaxe
A maior parte das linguagens usa chaves para agrupar pedaços de código (escopos). Por exemplo, em C tudo em um loop for é cercado por um ''{'' e um ''}'' . Mas não em Python. Identação é a chave. Tudo que estiver em um mesmo nível de identação fará parte do mesmo grupo de código:


*   C:

```c
for (i = 0; i < 10; i++){
    printf("%d\n", i)
}
printf("Fim!\n")

```
* Python:

````
for i in range(0,10):
  print(i)
print("Fim!")
````

## Listas
Listas são estruturas muito versáteis:

````
lista_de_compras = ["Leite", "Pão", "Queijo", "Presunto"]
print(lista_de_compras)
````

Para adicionar um item à lista:

````
lista_de_compras = lista_de_compras + ["Manteiga"]
print(lista_de_compras)
````

Para acessar apenas um item da lista:

````
print(lista_de_compras[2])
````

Para acessar um intervalo dentro de uma lista,  digamos o intervalo $\left[0, 2\right[$:

````
print(lista_de_compras[0:3])
````

Lembrando que listas não necessariamente devem conter o mesmo tipo de dado. Podemos ter listas como: 

````
lista = [2, "Pão", 400, "Presunto"]
print(lista)
````

## Loops for
Essa é a maior diferença da maioria das linguagens. Tradicionalmente, você tem uma variavel e a incrementa até que alguma condição seja falsa. Com loops `for` em Python, você tem uma lista e uma variavel e o interpretador itera através de cada valor da lista fazendo essa variável ser igual a esse valor. 

Exemplo:

````
for item in lista_de_compras:
  print("Preciso comprar " + item)
````

Se o objetivo for iterar sobre um intervalo de números, então deve-se utilizar a função `range(x, y)` (lembrando que o intervalo será $\left[x, y\right[$):

````
a = range(0,5)
for i in a:
  print(i)
````

## Funções

Funções são definidas através da declaração `def`, da seguinte forma:

````
def foo():
  a = 1
  b = 4
  print (a + b)
foo()
````

Também é possível ter mais de um nível de identação dentro de um `def`, da mesma maneira que você teria se estivesse usando chaves:

````
def foo():
  msg = "Essa é a mensagem número: "
  rep = 10
  for i in range(0, rep):
    print(msg + str(i))
  print("Fim!")
  
foo()
````

Parametros são passados da forma:

````
def foo(a, b):
  print(a + b)
foo(50, 40)
foo(10, 20)
````

## Declarações if


```python
if alguma condição:
  faça isso
```
Portanto:


````
x = 5
if x > 4:
  print("x é maior que 4!")
````

Também temos o `else`:

````
x = 5
if x > 4:
  print("x é maior que 4!")
else:
  print("x é menor ou igual a 4!")
````

E finalmente else if's, denotados por `elif`:

````
x = 4
if x > 5:
  print("x é maior que 5!")
elif x < 5:
  print("x é menor que 5!")
else:
  print("x é igual a 5!")
````

## Loops While
Loops `while` são do jeito que você os imaginava:


```python
while condição:
  faça isso
```

Exemplo:

````
a = 0
while a < 10:
  print(a)
  a = a + 1
````

## Strings
Strings podem ser concatenadas com o operador `+`:

````
a = "Hello"
b = " World!"
c = a + b
print(c)
````

Os operadores `=` e `==` funcionam como a maioria das linguagens:

````
a = "foob"
if a == "foo":
  print("a é igual a foo")
else:
  print("a não é igual a foo")
````

## Comparadores And e Or
Em python não existe && ou ||. Só existem `and` e `or`:

````
a = 1
b = 1
if (a == 1) and (b == 1):
  print("a e b são iguais a 1!")
elif (a == 1) or (b == 1):
  print("Ou a ou b é igual a 1!")
else:
  print("Nem a nem b é igual a 1!")
````
